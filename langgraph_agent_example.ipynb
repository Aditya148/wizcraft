# AgentViz Platform Integration Example

This notebook demonstrates how a Python-based agent (e.g., built with LangGraph, AutoGen, or CrewAI) can push data and visualizations directly to the AgentViz platform in real-time.

> **Prerequisites:** Ensure the AgentViz Gateway server is running locally on port 4000 before running these cells (`npm run server`).

```python
import requests
import json
import uuid
import time
from datetime import datetime

AGENT_GATEWAY_URL = "http://localhost:4000/api/push"

def send_to_dashboard(viz_type, data, config=None, summary=""):
    """
    Pushes data to the AgentViz platform to render a widget.
    """
    payload = {
        "id": f"resp-{uuid.uuid4()}",
        "agent_id": "langgraph-agent", 
        "timestamp": datetime.utcnow().isoformat() + "Z",
        "viz_type": viz_type,
        "viz_config": config or {},
        "data": data,
        "text_summary": summary
    }
    
    try:
        response = requests.post(
            AGENT_GATEWAY_URL, 
            json=payload,
            headers={"Content-Type": "application/json"}
        )
        if response.status_code == 200:
            print(f"✅ Successfully pushed '{viz_type}' widget to the dashboard!")
        else:
            print(f"⚠️ Server returned status code {response.status_code}: {response.text}")
    except Exception as e:
        print(f"❌ Failed to push to dashboard. Is the server running on port 4000? Error: {e}")
```

## 1. Pushing a KPI Grid
Let's say your agent just analyzed some general performance metrics. We can push them as KPI cards.

```python
kpi_data = [
    { "label": "Total Revenue", "value": "$2.4M", "change": "+12.5%", "trend": "up" },
    { "label": "Active Users", "value": "184K", "change": "+8.2%", "trend": "up" },
    { "label": "Churn Rate", "value": "4.1%", "change": "-1.1%", "trend": "down" }
]

send_to_dashboard(
    viz_type="kpi_grid",
    data=kpi_data,
    summary="I have analyzed the top line metrics. Revenue and user growth are trending positively."
)
```

## 2. Pushing a Bar Chart
Now the agent breaks down the revenue by specific product lines.

```python
bar_chart_data = [
    { "product": "Widget Pro X", "revenue": 284500, "growth": 0.15 },
    { "product": "CloudSync Plus", "revenue": 198000, "growth": 0.08 },
    { "product": "DataVault Enterprise", "revenue": 175000, "growth": -0.02 },
    { "product": "SensorMax IoT", "revenue": 156000, "growth": 0.22 }
]

send_to_dashboard(
    viz_type="chart",
    data=bar_chart_data,
    config={
        "title": "Revenue by Product Line",
        "chart_type": "bar",
        "x_axis": "product",
        "y_axis": "revenue"
    },
    summary="Widget Pro X is leading the product revenue board this quarter."
)
```

## 3. Pushing a Data Table
Sometimes raw numbers are needed. The agent can push a sortable, filterable data table.

```python
table_data = [
    { "id": "ORD-001", "status": "Delivered", "amount": 1200, "customer": "Acme Corp" },
    { "id": "ORD-002", "status": "Processing", "amount": 850, "customer": "Globex" },
    { "id": "ORD-003", "status": "Shipped", "amount": 2100, "customer": "Initech" },
    { "id": "ORD-004", "status": "Cancelled", "amount": 400, "customer": "Soylent" },
]

send_to_dashboard(
    viz_type="table",
    data=table_data,
    config={
        "title": "Recent High-Value Orders"
    },
    summary="Here are the recent orders exceeding standard thresholds."
)
```

## 4. Pushing a Geo Map
If your agent handles spatial data, it can render regional performance maps.

```python
geo_data = [
    { "region": "North America", "sales": 850000 },
    { "region": "Europe", "sales": 620000 },
    { "region": "Asia Pacific", "sales": 480000 },
    { "region": "Latin America", "sales": 210000 }
]

send_to_dashboard(
    viz_type="geo_map",
    data=geo_data,
    config={
        "title": "Global Sales Distribution",
        "name_field": "region",
        "value_field": "sales"
    },
    summary="North America remains the strongest region, accounting for the plurality of sales."
)
```

## 5. Statistical Intelligence: Correlation Matrix
Pushes a generated correlation matrix heatmap, showcasing ECharts native heatmapping for variables.

```python
correlation_data = [
    { "varX": "Age", "varY": "Age", "corr": 1.0 },
    { "varX": "Age", "varY": "Income", "corr": 0.65 },
    { "varX": "Age", "varY": "Tenure", "corr": 0.8 },
    { "varX": "Income", "varY": "Age", "corr": 0.65 },
    { "varX": "Income", "varY": "Income", "corr": 1.0 },
    { "varX": "Income", "varY": "Tenure", "corr": 0.45 },
    { "varX": "Tenure", "varY": "Age", "corr": 0.8 },
    { "varX": "Tenure", "varY": "Income", "corr": 0.45 },
    { "varX": "Tenure", "varY": "Tenure", "corr": 1.0 }
]

send_to_dashboard(
    viz_type="chart",
    data=correlation_data,
    config={
        "title": "Feature Correlation Matrix",
        "chart_type": "heatmap",
        "x_axis": "varX",
        "y_axis": "varY",
        "value": "corr"
    },
    summary="Strong positive correlation (r=0.8) detected between Age and Tenure."
)
```

## 6. Statistical Intelligence: Regression Scatter
Pushes a scatter plot representing a regression analysis snapshot.

```python
import random
scatter_data = [{ "Advertising X": x, "Sales Y": (x * 1.5) + random.randint(-20, 20) } for x in range(10, 100, 5)]

send_to_dashboard(
    viz_type="chart",
    data=scatter_data,
    config={
        "title": "Linear Regression: Adv vs. Sales",
        "chart_type": "scatter",
        "x_axis": "Advertising X",
        "y_axis": "Sales Y"
    },
    summary="R²=0.91 indicates strong predictive power of Advertising on Sales."
)
```
